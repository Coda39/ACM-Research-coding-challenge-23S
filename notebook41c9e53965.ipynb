```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
star_data = pd.read_csv('../input/star-dataset/6 class csv.csv', names=['Temperature', 'Luminosity', 'Radius', 'Abs_magnitude', 'Star_type', 'Star_color', 'Spectral_class']) # read in the star-dataset
# create variable names for the data
star_temperature_data = star_data["Temperature"] 
star_luminosity_data = star_data["Luminosity"] 
star_radius_data = star_data["Radius"] 
star_magnitude_data = star_data["Abs_magnitude"] 
star_type_data = star_data.Star_type
star_color_data = star_data.Star_color
star_class_data = star_data["Spectral_class"]

# creating two arrays of just the numerical data of both temperature and luminosity 
star_temperature_array = star_temperature_data.to_numpy()
star_temperature_array = star_temperature_array[1:].astype(float)

star_luminosity_array = star_luminosity_data.to_numpy()
star_luminosity_array = star_luminosity_array[1:].astype(float)

# use a scatter plot to plot the data and labeling the axis
plt.scatter(star_temperature_array, star_luminosity_array, c=star_temperature_array, cmap = 'inferno')
plt.xlabel('Temperature (K)')
plt.ylabel('Luminosity (L/Lo)')
plt.title("Hertzsprung–Russell diagram")
plt.yscale('log')
plt.xscale('log')
plt.colorbar(label="Temperature (K)")
plt.show()
```

The above code showcases a typical Hertzsprung-Russel Diagram using the temperature and luminosity data from the star-dataset.

I then wanted to see if I could match the size of the dots within the scatter plot to the radius of the stars.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
star_data = pd.read_csv('../input/star-dataset/6 class csv.csv', names=['Temperature', 'Luminosity', 'Radius', 'Abs_magnitude', 'Star_type', 'Star_color', 'Spectral_class'])
star_temperature_data = star_data["Temperature"]
star_luminosity_data = star_data["Luminosity"] 
star_radius_data = star_data["Radius"] 
star_magnitude_data = star_data["Abs_magnitude"] 
star_type_data = star_data.Star_type
star_color_data = star_data.Star_color
star_class_data = star_data["Spectral_class"]

star_temperature_array = star_temperature_data.to_numpy()
star_temperature_array = star_temperature_array[1:].astype(float)

star_luminosity_array = star_luminosity_data.to_numpy()
star_luminosity_array = star_luminosity_array[1:].astype(float)

# create an array of the radius data of similar structure to the previous arrays
star_radius_array = star_radius_data.to_numpy()
star_radius_array = star_radius_array[1:].astype(float)
sizes = star_radius_array

# pass the sizes array into the 's' parameter in the scatter function
plt.scatter(star_temperature_array, star_luminosity_array, s = sizes, c=star_temperature_array, cmap = 'inferno')
plt.xlabel('Temperature (K)')
plt.ylabel('Luminosity (L/Lo)')
plt.title("Hertzsprung–Russell diagram")
plt.yscale('log')
plt.xscale('log')
plt.colorbar(label="Temperature (K)")
plt.show()
```

The above showcases the different scatter points with their respective radius. This showcases the vast differences in size from the supergiants and giants to the smaller dwarf stars. 

As a final question I wondered if I could correlate the star color with the actual color within the dataset and not just use the temperature to determine the color.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
star_data = pd.read_csv('../input/star-dataset/6 class csv.csv', names=['Temperature', 'Luminosity', 'Radius', 'Abs_magnitude', 'Star_type', 'Star_color', 'Spectral_class'])
star_temperature_data = star_data["Temperature"]
star_luminosity_data = star_data["Luminosity"] 
star_radius_data = star_data["Radius"] 
star_magnitude_data = star_data["Abs_magnitude"] 
star_type_data = star_data.Star_type
star_color_data = star_data.Star_color
star_class_data = star_data["Spectral_class"]

star_temperature_array = star_temperature_data.to_numpy()
star_temperature_array = star_temperature_array[1:].astype(float)

star_luminosity_array = star_luminosity_data.to_numpy()
star_luminosity_array = star_luminosity_array[1:].astype(float)

star_radius_array = star_radius_data.to_numpy()
star_radius_array = star_radius_array[1:].astype(float)
sizes = star_radius_array

# create an array of the star color data as datatype string
star_color_array = star_color_data.to_numpy()
star_color_array = star_color_array[1:]
star_color_array = star_color_array.astype(str)

# map each value in the string array to a numerical value 
color_map ={'Blue': 0, 'Blue White': 1, 'Red': 2, 'Orange': 3, 'White': 4, 'yellow-white': 5, 'Whitish': 6, 'Yellowish': 7, 'Orange-Red': 8, 'Pale yellow orange': 9, 'Yellowish White': 10, 'Blue white': 11, 'Blue-white': 12, 'White-Yellow': 13, 'white': 14, 'Blue ': 15, 'yellowish': 16, 'Blue White': 17, 'Blue white ': 18, 'Blue-White': 19}
# create a numerical array that corresponds to each value in the string array using the created color map
colors = np.array([color_map[s] for s in star_color_array])

plt.scatter(star_temperature_array, star_luminosity_array, s = sizes, c=colors, cmap = 'inferno')
plt.xlabel('Temperature (K)')
plt.ylabel('Luminosity (L/Lo)')
plt.title("Hertzsprung–Russell diagram")
plt.yscale('log')
plt.xscale('log')
plt.colorbar(label="Temperature (K)")
plt.show()
```
